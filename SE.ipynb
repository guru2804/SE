# ComparingFunctions

```python
class FunctionComparer():
    def __init__(self,f1,f2):
        self.file1=f1.getFileName()
        self.file2=f2.getFileName()
    def getSimilarFunctions(self):
        set1=set(self.Pf1)
    	set2=set(self.Pf2)
     	return (set1&set2)
    def getDiffFunctions(self):
        set1=set(self.Pf1)
    	set2=set(self.Pf2)
     	return (set1.difference(set1&set2),set2.difference(set1&set2))


    def compareFunctions(self):

        x=open(self.file1,"r")
        y=open(self.file2,"r")
        k=x.read()
        f=y.read()

        a=(k.split("int main"))
        b=(f.split("int main"))
        
        s=a[0].split("\n")
        p=b[0].split("\n")
        
        
        res=[]
        res1=[]
        
        for i in s:
            x = re.search("^int.*()", i)
            y = re.search("^float.*()", i)
            z = re.search("^bool.*()", i)
            q = re.search("^void.*()", i)
            r = re.search("^char.*()", i)
            if(x):
                m=x.span()
                res.append(i[m[0]:m[1]])
            if(y):
                m=y.span()
                res.append(i[m[0]:m[1]])
            if(z):
                m=z.span()
                res.append(i[m[0]:m[1]])
            if(q):
                m=q.span()
                res.append(i[m[0]:m[1]])
            if(r):
                m=r.span()
                res.append(i[m[0]:m[1]])
        
        for i in p:
            x = re.search("^int.*()", i)
            y = re.search("^float.*()", i)
            z = re.search("^bool.*()", i)
            q = re.search("^void.*()", i)
            r = re.search("^char.*()", i)
            if(x):
                m=x.span()
                res1.append(i[m[0]:m[1]])
            if(y):
                m=y.span()
                res1.append(i[m[0]:m[1]])
            if(z):
                m=z.span()
                res1.append(i[m[0]:m[1]])
            if(q):
                m=q.span()
                res1.append(i[m[0]:m[1]])
            if(r):
                m=r.span()
                res1.append(i[m[0]:m[1]])
                
            
            
    
        self.Pf1=res
        self.Pf2=res1
    
        return (self.getSimilarFunctions(),self.getDiffFunctions())

       
```

# ComparingClasses

```python
class ClassComparer():
    def __init__(self,f1,f2):
     	self.file1=f1.getFileName()
    	self.file2=f2.getFileName()
    def getSimilarClasses(self):
    	set1=set(self.Pf1)
    	set2=set(self.Pf2)
     	return (set1&set2)
    	
    def getDiffClasses(self):
    	set1=set(self.Pf1)
    	set2=set(self.Pf2)
    	return (set1.difference(set1&set2),set2.difference(set1&set2))
  


    def compareClass(self):

        x=open(self.file1,"r")
        y=open(self.file2,"r")
        k=x.read()
        f=y.read()
        s=(k.split("\n"))
        p=(f.split("\n"))
        subs="class"
        res=[i for i in s if subs in i]
        res1=[i for i in p if subs in i]
        self.Pf1=res
        self.Pf2=res1
       
        return (self.getSimilarClasses(),self.getDiffClasses())
```

# Printing Functions

```python
def printers(p,q,r,s):
    print("\n")
    print("Similarity in Both Files\n")
    length=len(p)
    print("Number of Classes:%d"%(length))
    
    print("\n")
    for i in p:
        print(i)
    print("\n")
    
    length=len(q)
    print("Number of Functions:%d"%(length))
    
    print("\n")
    for i in q:
        print(i)
    print("\n")
    
    length=len(r)
    print("Number of Variables:%d"%(length))
    
    print("\n")
    for i in r:
        print(i)
    print("\n")
    
    length=len(s)
    print("Number of Structures:%d"%(length))
    
    print("\n")
    for i in s:
        print(i)
    print("\n")
    
        
```

```python
print("Welcome To The Source Code Comparison Tool\n")
x=input("Enter the Path of File 1:")
file1=File(x)
y=input("Enter the Path of File 2:")
file2=File(y)

print("Path of File 1:"+file1.getFileName())
print("Path of File 2:"+file2.getFileName())

ClassVar=ClassComparer(file1,file2)
FunctionVar=FunctionComparer(file1,file2)
VariableVar=VariableComparer(file1,file2)
StructVar=StructComparer(file1,file2)
simclass,diffclass=ClassVar.compareClass()
simfunc,difffunc=FunctionVar.compareFunctions()
simvar,diffvar=VariableVar.compareVariables()
simstruct,diffstruct=StructVar.compareStruct()

printers(simclass,simfunc,simvar,simstruct)
printeru(diffclass,difffunc,diffvar,diffstruct)
```
